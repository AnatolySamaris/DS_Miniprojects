Эта функция считает суммы чисел из диапазона **[a, b)** и их "отражений", т.е. перевёрнутых чисел. На выходе получаем небольшую статистику и графики:
- С поведением функций f(x) = x + \x, g(x) = \x, q(x) = x;
- С распределением числа повторений оригинальных сумм.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def check_range(start, end):
  nums = [i + int(str(i)[::-1]) for i in range(start, end)]
  reflected = [int(str(i)[::-1]) for i in range(start, end)]

  sum_count = {}
  for i in nums:
    try:
      sum_count[i] += 1
    except:
      sum_count[i] = 1
    
  sum_amount = len(sum_count)

  sums = sorted(sum_count.keys())
  sums_count = [sum_count[i] for i in sums]

  print(f'РАСЧЁТЫ НА ДИАПАЗОНЕ [{start}, {end}):')
  print('Оригинальных сумм:', sum_amount)
  print('Максимальное число повторений суммы:', max(sum_count.values()))
  print('Минимальное число повторений суммы:', min(sum_count.values()))

  plt.figure(figsize=(16, 8))

  plt.subplot(1, 2, 1)
  plt.plot(range(start, end), nums, label='Сумма')
  plt.plot(range(start, end), reflected, label='Отражения')
  plt.plot(range(start, end), range(start, end), label='Числа')
  plt.legend()
  plt.title('Поведение чисел, их отражений и суммы чисел и отражений')

  plt.subplot(1, 2, 2)
  plt.plot(sums, sums_count)
  plt.title('Распределение повторений оригинальных сумм')
```

```python
check_range(10, 100)
```

```python
check_range(100, 1000)
```

```python
check_range(1000, 10000)
```

```python
check_range(10000, 100000)
```

```python
check_range(100000, 1000000)
```
